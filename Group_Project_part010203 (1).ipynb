 Good to know since xG is a metric in the data pulled:
 In football, xG (Expected Goals) is a performance metric that measures the
 probability of a shot resulting in a goal. Each shot is assigned a score
 between 0 and 1, representing its likelihood of scoring, based on factors
 like shot distance, angle, player position, and shot type. By summing the
 xG values for all shots in a match, it shows how many goals a team should
 have scored.

## Background
Meaning of the columns:
npg	npxG	xGChain	xGBuildup

*   **npg - Non-Penalty Goals**
Why exclude penalties? Because they’re relatively “easy” and don’t reflect open-play performance.

Use case: Compare attackers fairly — a player with 20 goals including 10 penalties might be less impressive than one with 15 all from open play.

*   **npgX  - Non-Penalty Expected Goals**
Sum of expected goals excluding penalties

xG = the probability of a shot becoming a goal based on distance, angle, body part, defensive pressure, etc.

npxG removes penalties (which are almost always ~0.76 xG) to show open-play shot quality.

Why it’s important: If a striker has 12 npg but npxG of 9.0 → he’s overperforming (finishing very well).
If npxG is 11.5 but only 7 npg → underperforming (poor finishing or bad luck).

*  **xGChain - Expected goals chain** (mid field and attack)
It measures how much a player contributes to dangerous attacks through passing, movement, buildup, not just shooting.

Example: Player A doesn’t shoot much but is involved in every build-up leading to high-quality chances.
His xGChain might be very high even if his personal xG is low → he’s crucial in creating those chances.

*   **xGBuildup - Expected Goals Buildup** (defensive mid fielders and defenders)
Similar to xGChain, but excludes shots, key passes, and assists.

It’s purely about involvement in the earlier parts of the attack — passing, ball progression, linking play.

Example: A deep-lying midfielder or centre-back might have a high xGBuildup even if they never shoot or assist — showing their importance in starting attacking moves.



**How to use them in analysis:**

> Finishing skill: Compare npg vs. npxG

> Chance involvement: Sort players by xGChain

> Deep playmaking / buildup role: Sort by xGBuildup

For example:

> A striker’s job: high npg, slightly > npxG

> A creative midfielder: modest npg, high xGChain

> A deep midfielder: low npg, high xGBuildup

```python
# Multi-league, multi-season pull
# Leagues: Premier League, La Liga, Ligue 1, Serie A, Bundesliga
# Keeps it simple, JSON-encodes nested fields, and stores per season in clean folders.

# Map your human names -> Understat accepted keys
LEAGUES = {
    "Premier_League": "EPL",
    "La_Liga": "La_Liga",
    "Ligue_1": "Ligue_1",
    "Serie_A": "Serie_A",
    "Bundesliga": "Bundesliga",
}

# Choose seasons (Understat uses start year strings; "2020" -> 2020/21)
SEASONS = [str(y) for y in range(2019, 2025)]  # adjust as you like (5–10 years is fine)

BASE_DIR = Path("data")  # all output goes here

def to_df(records):
    return pd.DataFrame.from_records(records) if records else pd.DataFrame()

def jsonify_objects(df: pd.DataFrame) -> pd.DataFrame:
    """Convert dict/list/tuple cells to JSON strings so CSV/Parquet save cleanly."""
    if df.empty:
        return df
    df = df.copy()
    for c in df.select_dtypes(include=["object"]).columns:
        df[c] = df[c].apply(
            lambda v: json.dumps(v, ensure_ascii=False) if isinstance(v, (dict, list, tuple)) else v
        )
    return df

def save(df: pd.DataFrame, base_path: Path, csv=True, parquet=True):
    """Save a DataFrame to CSV/Parquet at base_path (without extension)."""
    base_path.parent.mkdir(parents=True, exist_ok=True)
    if df.empty:
        print(f"[warn] {base_path}: empty, nothing saved.")
        return
    out = jsonify_objects(df)
    if csv:
        out.to_csv(base_path.with_suffix(".csv"), index=False)
    if parquet:
        try:
            out.to_parquet(base_path.with_suffix(".parquet"), index=False)
        except Exception as e:
            print(f"[warn] Parquet save failed for {base_path.name}: {e}")
    print(f"[ok] saved {base_path.name} (in {base_path.parent})")

with UnderstatClient() as us:
    for nice_name, understat_key in LEAGUES.items():
        print(f"\n================ {nice_name} ================")

        # Accumulators for combined files (per league)
        all_players, all_matches, all_teams = [], [], []

        for season in SEASONS:
            print(f"\n--- Pulling {nice_name} {season} ---")
            # Pull league data
            players_raw = us.league(understat_key).get_player_data(season=season)
            matches_raw = us.league(understat_key).get_match_data(season=season)
            teams_raw   = us.league(understat_key).get_team_data(season=season)

            # To DataFrames
            df_players = to_df(players_raw)
            df_matches = to_df(matches_raw)
            df_teams   = to_df(teams_raw)

            # Tag for clarity in downstream analysis
            for df in (df_players, df_matches, df_teams):
                if not df.empty:
                    df["league"] = nice_name
                    df["season"] = season  # start-year format

            # Save per-season into data/<league>/<season>/
            season_dir = BASE_DIR / nice_name / season
            save(df_players, season_dir / "players")
            save(df_matches, season_dir / "matches")
            save(df_teams,   season_dir / "teams")

            # Collect for combined
            if not df_players.empty: all_players.append(df_players)
            if not df_matches.empty: all_matches.append(df_matches)
            if not df_teams.empty:   all_teams.append(df_teams)

            # Gentle pause (keeps things polite if you later extend to heavier endpoints)
            sleep(0.6)

        # Save combined per-league across seasons into data/<league>/combined/
        combined_dir = BASE_DIR / nice_name / "combined"
        if all_players:
            save(pd.concat(all_players, ignore_index=True), combined_dir / "players_allseasons")
        if all_matches:
            save(pd.concat(all_matches, ignore_index=True), combined_dir / "matches_allseasons")
        if all_teams:
            save(pd.concat(all_teams, ignore_index=True), combined_dir / "teams_allseasons")

print("\nDone.")
```

```python
def mergeallData(understat_transfermarkt, fbref, nameofcsv):
    nameofcsv = nameofcsv
    understat_transfermarkt= understat_transfermarkt.copy()
    fbref = fbref.copy()
    understat_transfermarkt.rename(columns={'player_name': 'Player'}, inplace=True)
    fbref.rename(columns={'Ast':'assists'},inplace=True)
    fbref['assists'] = fbref['assists'].astype(float)
    understat_transfermarkt['assists'] = understat_transfermarkt['assists'].astype(float)
    df_allmerge = pd.merge(
        fbref,
        understat_transfermarkt,
        on=['Player','assists'],
        how='inner' # left join since keeper is limited in number
    )
    #df_allmerge.info()
    df_allmerge.to_csv((f'{nameofcsv}.csv'), index=False)
    return df_allmerge
```

Important and relevant to the Analysis are the normalized p90 Statistics. These can be calculated by the dividing statistics by the number of 90 Minute matches played (90's).

Calculating the 90s for all the years and saving them as csv'S


```python
def normalizep90(all_data, nameof90scsv):
    nameof90scsv = nameof90scsv
    df_90s = all_data.copy()
    df_90s.drop(['SCA90','GA90'], axis=1, inplace=True) # We want to drop these values
    exclude_cols = ['Age', 'Born', '90s', 'Tkl%_challenges','Cmp%','short_Cmp%','medium_Cmp%','long_Cmp%','Save%','CS%','Save%_Pen', 'games', 'time', 'market_value_in_eur_y']
    numerical_cols = df_90s.select_dtypes(include=['int64', 'float64']).columns #to make sure to include the rest of the numerical columns
    cols_to_divide = [col for col in numerical_cols if col not in exclude_cols]# select the Columns to divide by '90s'
    for col in cols_to_divide:
        df_90s[col] = df_90s[col].divide(df_90s['90s']).replace([np.inf, -np.inf], np.nan) # Divide selected columns by '90s', division by zero is handled by replacing columns where 90s value is 0 to Nan
    df_90s.to_csv((f'{nameof90scsv}.csv'), index=False)
    return df_90s
```

```python
#!pip install adjustText #Only run this once

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# exploding the positions into "Role" so that players with more than one position can be considered for both
def prepare_positions(df, pos_col="Pos"):
    out = df.copy()
    out["Role"] = out[pos_col].astype(str).str.split(",")
    out = out.explode("Role")
    out["Role"] = out["Role"].str.strip()
    return out

# scatter plot function
def plot_position_scatter(
    df, pos, x, y,
    leagues=None,       # e.g., ['eng Premier League','es La Liga','it Serie A','fr Ligue 1','de Bundesliga']
    min_90s=15,         # here we set the default minimum amount of games that players have to play to qualify for our Analysis
    top_n=5,
    title=None,
    pos_col="Pos",
    color_by_comp=True,
    min_market_value = None,
    max_market_value = None,
    age_min = None,
    age_max = None,
    squad_name = None,
):

# Keep only the rows for the chosen Position and (optionally) chosen leagues, market value range as well asage range and the Squad (the team). It should be noted that this won't work if league filter is used and a squad name is given that is not part of the league
    data = prepare_positions(df, pos_col=pos_col)
    data = data[data["Role"] == pos]
    if leagues:
        data = data[data["Comp"].isin(leagues)]
    if "market_value_in_eur_y" in data.columns:
        if min_market_value is not None:
            data = data[data["market_value_in_eur_y"] >= min_market_value]
        if max_market_value is not None:
            data = data[data["market_value_in_eur_y"] <= max_market_value]
    if "Age" in data.columns:
        if age_min is not None:
            data = data[data["Age"] >= age_min]
        if age_max is not None:
            data = data[data["Age"] <= age_max]
    if "Squad" in data.columns:
        if squad_name is not None:
          data = data[data["Squad"] == squad_name]



# Keep only the columns we need & drop NAs
    cols = ["Player","Comp","90s",x,y]
    cols = [c for c in cols if c in data.columns]
    data = data[cols].dropna(subset=[x,y]).copy()
    data = data[data["90s"] >= min_90s]
    if data.empty:
        print(f"No rows for {pos} with {x} vs {y} after filters."); return

    # marker size ~ 90s (bounded)
    '''
    I clip the top/bottom 5% so one iron-man doesn’t dwarf everyone else. Then linearly scale to a pleasant size range.
    Why: Size should communicate playing time but stay readable. - ALTHOUGH IF YOU WANT TO CHANGE THIS, VERY UNDERSTANDABLE, IT COULD BE IMPORTANT TO KEEP.
    '''
    z = data["90s"]#.clip(data["90s"].quantile(0.05), data["90s"].quantile(0.95)) # trim extreme outliers
    s = 40 + 160 * (z - z.min()) / (z.max() - z.min() + 1e-9)

    # colors by league
    if color_by_comp and "Comp" in data.columns:
        codes, uniq = pd.factorize(data["Comp"]) # maps each league to an integer that plugs into a colormap
        c, cmap = codes, plt.cm.tab20
    else:
        c, cmap, uniq = "tab:blue", None, []

    plt.figure(figsize=(8,6))
    plt.scatter(data[x], data[y], s=s, c=c, cmap=cmap, alpha=0.35, edgecolor="none")
    plt.xlabel(x); plt.ylabel(y)
    plt.title(title or f"{pos}: {y} vs {x}")
    plt.grid(True, ls="--", lw=0.4, alpha=0.4)

    # label top-N by simple impact score (x*y)
    stars = data.assign(score=data[x]*data[y]).nlargest(top_n, "score") # You can change the score easily (e.g., 0.6*x + 0.4*y, or percentiles).
    try:
        from adjustText import adjust_text
        texts = [plt.text(r[x], r[y], r["Player"], fontsize=8) for _, r in stars.iterrows()] # If adjustText is installed, it nudges labels apart and draws subtle connectors.
        adjust_text(texts, arrowprops=dict(arrowstyle='-', color='gray', lw=0.5))
    except Exception:
        for _, r in stars.iterrows():
            jx, jy = np.random.uniform(-0.003,0.003,2)
            plt.annotate(r["Player"], (r[x]+jx, r[y]+jy), xytext=(4,4), # If not available, the code falls back to a tiny random jitter
                         textcoords="offset points", fontsize=8, alpha=0.95)

    # simple legend for leagues
    if color_by_comp and len(uniq) > 0:
        handles = [plt.Line2D([],[],marker='o',ls='',label=lab,
                              markerfacecolor=plt.cm.tab20(i%20), alpha=0.6)
                   for i, lab in enumerate(uniq)]
        plt.legend(handles=handles, title="League", frameon=False, fontsize=8, loc="best")

    plt.tight_layout()
    plt.show()
```

```python
def add_position_percentiles(df: pd.DataFrame, position: str, metrics: list) -> pd.DataFrame:

    position_col = "Pos"
    minutes_col = "90s"

    # Error Messages
    if position_col not in df.columns:
        raise KeyError("Column 'Pos' not found in DataFrame.")
    if minutes_col not in df.columns:
        raise KeyError("Column '90s' not found in DataFrame.")
    valid_metrics = [m for m in metrics if m in df.columns]
    if not valid_metrics:
        raise ValueError("None of the provided metrics exist in the DataFrame.")

    # Copy the Dataframe to work with it
    df = df.copy()

    # We also want to work with players that have multiple positions included
    pos_series = (
        df[position_col]
        .astype(str)
        .str.replace(r"\s+", " ", regex=True)
        .str.strip()
        .str.split(r"[;/,]")
    )
    exploded = df.assign(_position=pos_series).explode("_position")
    exploded["_position"] = exploded["_position"].astype(str).str.strip()
    exploded = exploded[exploded["_position"] != ""]

    # We wanna remove all the players that did not play
    exploded = exploded[exploded[minutes_col] > 0]

    # Filter to the requested position
    target_pos = position.strip()
    mask = exploded["_position"] == target_pos

    # Calculating the Percentiles
    for m in valid_metrics:
        ranks = exploded.loc[mask, m].rank(method="average", pct=True) * 100
        exploded.loc[mask, f"{m}_percentile"] = ranks

    # We wanna only keep that position
    out = exploded.loc[mask].copy()

    # We want to add the other relevant columns to our list of columns
    id_cols = ["Player", "Nation", "Squad", "Age", "Comp"]
    id_cols = [c for c in id_cols if c in out.columns]

    percentile_cols = [f"{m}_percentile" for m in valid_metrics]
    final_cols = id_cols + [position_col, "_position", minutes_col] + valid_metrics + percentile_cols
    final_cols = [c for c in final_cols if c in out.columns]

    return out[final_cols]
```

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

sns.set_theme(style="whitegrid")


def plot_radar(players, df, metrics):
    # Filter relevant players
    selected = df[df["Player"].isin(players)].copy()
    if selected.empty:
        raise ValueError("None of the selected players were found in the DataFrame.")

    # Radar chart setup
    N = len(metrics)
    angles = np.linspace(0, 2 * np.pi, N, endpoint=False).tolist()
    angles += angles[:1]

    # Initialize plot
    plt.figure(figsize=(7, 7))
    ax = plt.subplot(111, polar=True)

    # Distinct colors for each player
    palette = sns.color_palette("husl", len(players))

    for i, player in enumerate(players):
        if player not in selected["Player"].values:
            continue

        row = selected[selected["Player"] == player].iloc[0]
        values = [row[m + "_percentile"] for m in metrics]
        values += values[:1]

        ax.plot(angles, values, linewidth=2, label=player, color=palette[i])
        ax.fill(angles, values, alpha=0.25, color=palette[i])

    # Labels and grid
    plt.xticks(angles[:-1], metrics, color='grey', size=9)
    ax.set_yticklabels([])# we do not want any y tick labels

    # Title & legend
    title_str = " vs ".join(players)
    plt.title(f"{title_str} — Percentile Comparison", size=14, y=1.08)
    plt.legend(loc='upper right', bbox_to_anchor=(1.2, 1.1))
    plt.tight_layout()
    plt.show()


plot_radar(["Robert Lewandowski", "Harry Kane"], df_fw, fw_metrics)
```

```python
def find_similar_players(
    df: pd.DataFrame,
    player: str,
    metrics: list[str] | None = None,
    k: int = 5,
    min_90s: float = 10,# float if we wish to have at least 9.5 90s played
    same_position: bool = True
):
    position_col = "Pos"
    minutes_col = "90s"

    if player not in df["Player"].values:
        raise ValueError(f"Player '{player}' not found in DataFrame.")

    # Determine metric columns
    if metrics is None:
        pct_cols = [c for c in df.columns if c.endswith("_percentile")]
        metrics = [c[:-11] for c in pct_cols]  # strip "_percentile"
    pct_cols = [f"{m}_percentile" for m in metrics]

    missing = [c for c in pct_cols if c not in df.columns]
    if missing:
        raise KeyError(f"Missing percentile columns: {missing}")

    # Base filters: valid minutes, has all percentiles
    base = df[(df[minutes_col] > min_90s)].copy()
    base = base.dropna(subset=pct_cols)

    # Identify the query player's row and positions
    q = base[base["Player"] == player]
    if q.empty:
        raise ValueError(f"Player '{player}' fails filters (e.g., minutes <= {min_90s} or missing percentiles).")
    q_row = q.iloc[0]

    # Exclude the query player from candidate set
    candidates = base[base["Player"] != player].copy()
    if candidates.empty:
        return pd.DataFrame(columns=["Player", position_col, minutes_col, "distance"])

    # Build vectors for euclidian distance calculation
    Q = q_row[pct_cols].to_numpy(dtype=float)
    X = candidates[pct_cols].to_numpy(dtype=float)

    # Euclidian Distance calculation
    dists = np.linalg.norm(X - Q, axis=1)

    candidates = candidates.copy()
    candidates["distance"] = dists

    # Sort ascending and take top-k-players
    out = (
        candidates
        .sort_values("distance", ascending=True)
        .loc[:, ["Player", position_col, minutes_col, "distance"] + pct_cols]
        .head(k)
    )

    return out
player_to_compare = 'Harry Kane'
similar_players = find_similar_players(df_fw_n,player_to_compare, fw_metrics, 4)
#all_players = similar_players["Player"].tolist()
#similar_players["Player"].tolist().append(player_to_compare)

similar_player_list = similar_players["Player"].tolist()
print(similar_player_list)
all_player_list = similar_player_list+ [player_to_compare]
print(all_player_list)
plot_radar(all_player_list, df_fw, fw_metrics)
```
